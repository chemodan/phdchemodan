The data are grouped into categories of:

1. Less than a high school diploma
2. High school diploma only
3. Some college or associate's degree
4. Some college (1-3 years)
5. Bachelor's degree or higher
6. Percent of adults with less than a high school diploma
7. Percent of adults with high school diploma only
8. Percent of adults completing some college or associate's degree
9. Percent of adults completing some college (1-3 years)
10. Percent of adults with a bachelor's degree or higher

We shall rename these categories into:

1. "lt_high_school_diploma"
2. "high_school_diploma"
3. "some_college_or_assoc_degree"
4. "some_college__1_3_yrs"
5. "bachelors_degree_plus"
6. "pct__lt_high_school_diploma
7. "pct__high_school_diploma"
8. "pct__some_college_or_assoc_degree"
9. "pct__some_college__1_3_yrs"
10. "pct__bachelors_degree_plus"

```python
def rename_columns(counties: pd.DataFrame) -> pd.DataFrame:
    cols = list(counties.columns)
    
    for cc in range(len(cols)):
        if "less than a high school diploma" in cols[cc].lower():
            cols[cc] = cols[cc].replace("Less than a high school diploma", "lt_high_school_diploma")
            cols[cc] = cols[cc].replace("less than a high school diploma", "lt_high_school_diploma")
            
        if "high school diploma only" in cols[cc].lower():
            cols[cc] = cols[cc].replace("High school diploma only", "high_school_diploma")
            cols[cc] = cols[cc].replace("a high school diploma only", "high_school_diploma")
            
        if "some college or associate's degree" in cols[cc].lower():
            cols[cc] = cols[cc].replace("Some college or associate's degree", "some_college_or_assoc_degree")
            cols[cc] = cols[cc].replace("some college or associate's degree", "some_college_or_assoc_degree")
            
        if "some college (1-3 years)" in cols[cc].lower():
            cols[cc] = cols[cc].replace("Some college (1-3 years)", "some_college_or_assoc_degree")
            cols[cc] = cols[cc].replace("some college (1-3 years)", "some_college_or_assoc_degree")
            
        if "bachelor's degree or higher" in cols[cc].lower():
            cols[cc] = cols[cc].replace("Bachelor's degree or higher", "bachelors_degree_plus")
            cols[cc] = cols[cc].replace("a bachelor's degree or higher", "bachelors_degree_plus")
        
        if "Percent of adults with " in cols[cc]:
            cols[cc] = cols[cc].replace("Percent of adults with ", "pct__")
        if "Percent of adults completing " in cols[cc]:
            cols[cc] = cols[cc].replace("Percent of adults completing ", "pct__")
            
        cols[cc] = cols[cc].replace(", ", "__")
        cols[cc] = cols[cc].replace("-", "_")
            
    counties.columns = cols
    
    return counties
```

```python
dfReady = rename_columns(counties.copy())
dfReady.head()
```

```python
myCols = [cc for cc in dfReady.columns if "Code" not in cc]
cols = ['state', 'county', 'county_state']
for cc in myCols:
    if cc not in cols:
        cols.append(cc)
myEdDF = dfReady[cols].copy()
myEdDF
```

```python
def meltEdDF(myEdDF: pd.DataFrame) -> pd.DataFrame:
    myEdDFlong = myEdDF.melt(id_vars=["state", "county", "county_state"])

    edPctLong = myEdDFlong.loc[myEdDFlong["variable"].str.contains("pct__")].copy()
    edPctLong
    edPctLong[["pct", "metric", "years"]] = edPctLong["variable"].str.split("__", expand=True)
    edPctLong["metric"] = edPctLong["pct"] + "__" + edPctLong["metric"]
    del edPctLong["pct"]
    del edPctLong["variable"]
    edPctLong.head(10)

    edCntLong = myEdDFlong.loc[~myEdDFlong["variable"].str.contains("pct__")].copy()
    edCntLong[["metric", "years"]] = edCntLong["variable"].str.split("__", expand=True)
    del edCntLong["variable"]
    edCntLong.head(10)

    myEdDFlong = pd.concat([edPctLong, edCntLong]).reset_index(drop=True)
    
    return myEdDFlong
```
